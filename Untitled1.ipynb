```python
# Define the two datasets
dataset1 = [1, 2, 3, 4, 5]
dataset2 = [6, 7, 8, 9, 10]

# Perform multiplication on the two datasets
result_multiplication = [a * b for a, b in zip(dataset1, dataset2)]
print("Result of multiplication: ", result_multiplication)

# Perform division on the two datasets
result_division = [a / b for a, b in zip(dataset1, dataset2)]
print("Result of division: ", result_division)

# Perform power operation on the two datasets
result_power = [a ** b for a, b in zip(dataset1, dataset2)]
print("Result of power operation: ", result_power)
```

```python
import numpy as np

# Define an array
array = np.array([1, 2, 3, 4, 5])

# Calculate the sine of the array
result_sin = np.sin(array)
print("The sine of the array is: ", result_sin)

# Calculate the natural logarithm of the array
result_log = np.log(array)
print("The natural logarithm of the array is: ", result_log)

# Calculate the base-2 logarithm of the array
result_log2 = np.log2(array)
print("The base-2 logarithm of the array is: ", result_log2)

# Calculate the exponential of the array
result_exp = np.exp(array)
print("The exponential of the array is: ", result_exp)
```

```python
import matplotlib.pyplot as plt
categories = ['Category A', 'Category B', 'Category C', 'Category D']
values = [15, 30, 25, 40]
plt.bar(categories, values, color='blue')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Simple Bar Chart')
plt.show()
```
